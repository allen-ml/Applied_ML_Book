Sorting

Use[::-1] for descending sort


```
ages=np.array([22,23,15,5,13])
sorted_ages=np.sort(ages)

print("ages",ages)
print("\n")

print("sorted ages",sorted_ages)
print("\n")

ages.sort()
print(ages)

ages=np.array([22,23,15,5,13])
print("ages",ages)
print("\n")
print("indices",ages.argsort())
print("\n")
print(ages[ages.argsort()])
print("\n")
#Descneding
indexReversed=np.argsort(ages)[::-1]
print(indexReversed)
print(ages[indexReversed])
```

```
import matplotlib.pyplot as plt
# represents the heights of a group of people in meters
heights = [[1.6], [1.65], [1.7], [1.73], [1.8]]
# represents the weights of a group of people in kgs
weights = [[60], [65], [72.3], [75], [80]]
plt.plot(heights, weights, 'k.')

# test data
heights_test = [[1.58], [1.62], [1.69], [1.76], [1.82]]
weights_test = [[58], [63], [72], [73], [85]]

from sklearn.linear_model import LinearRegression
 
# Create and fit the model
model = LinearRegression()
model.fit(X=heights, y=weights)
# make prediction
predictedWeight = model.predict(heights_test)
print(predictedWeight)

# plot the regression line
plt.plot(heights_test, model.predict(heights_test), color='r')
plt.show()

#Getting the Gradient and Intercept of the Linear Regression Line

#While you can get the y-intercept by predicting the weight if the height is 0:
round(model.predict([[0]])[0][0],2)   # -104.75
# the model object provides the answer directly through the intercept_ property:
print(round(model.intercept_[0],2))   # -104.75
#Using the model object, you can also get the gradient of the linear regression line through the coef_ property:
print(round(model.coef_[0][0],2))    # 103.31
# using scikit-learn to calculate r-squared
print('R-squared: %.4f' % model.score(heights_test,weights_test))
 
plt.title('Weights plotted against heights')
plt.xlabel('Heights in meters')
plt.ylabel('Weights in kilograms')
 
plt.plot(heights, weights, 'k.')
 
# axis range for x and y
plt.axis([1.5, 2.3, 50, 130])
plt.grid(True)
```

In Tukey Fences, outliers are values that are as follows:
 ■ Less than Q1 – (1.5 × IQR), or
 ■ More than Q3 + (1.5 × IQR)

 The outliers_iqr() function returns a tuple of which the frst element is an 
array of indices of those rows that have outlier values.

```
import numpy as np
 
def outliers_iqr(data):
    q1, q3 = np.percentile(data, [25, 75])
    iqr = q3 - q1
    lower_bound = q1 - (iqr * 1.5)
    upper_bound = q3 + (iqr * 1.5)
    #The np.where() function returns the location of items satisfying the conditions.
    return np.where((data > upper_bound) | (data < lower_bound))

import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/allen-ml/ML-Data/master/galton-data.csv")
print(df.head())
print("\n")
print(outliers_iqr(df.height))

print("Outliers using outliers_iqr()")
print("=============================")
for i in outliers_iqr(df.height)[0]:
    print(df[i:i+1])
print("\n")
```

**Z-Score**
The second method for determining outliers is to use the Z-score method. A 
Z-score indicates how many standard deviations a data point is from the mean. 
The Z-score has the following formula:
Z=(X-moo)/sigma
Any Z-score greater than 3 or less than –3 is considered to be an outlier

```
#Any Z-score greater than 3 or less than –3 is considered to be an outlier
def outliers_z_score(data): #data can be a column in your data
    threshold = 10
    mean = np.mean(data)
    std = np.std(data)
    z_scores = [(y - mean) / std for y in data]
    return np.where(np.abs(z_scores) > threshold)

print(outliers_iqr(df1.ReceivedQuantity))
print("Outliers using outliers_z_score()")
print("=================================")
for i in outliers_z_score(df1.SoldQuantity)[0]:
    print(df1.SoldQuantity[i:i+1])
print()
'''ReceivedQuantity    0
SoldQuantity        0
EndQuantity         0
LatestOrder         0
StockedOut          0
GroupID             0
MissedSales         0
'''
```
